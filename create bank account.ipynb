```python
class Account:
  
    def __init__(bank, account_number, account_balance, account_holder, account_deposit):
        bank.account_number = account_number
        bank.account_balance = account_balance
        bank.account_holder = account_holder
        bank.account_deposit = account_deposit
    
    def deposit(bank, amount):
     bank.account_balance += amount

    def withdraw(bank, amount):
     if amount > bank.account_balance:
       print('Insufficient funds')

     else:
       bank.account_balance -= amount
      


    def check_balance(bank):
     return (f'Your current balance is {bank.account_balance}')
    
my_account = Account("007878787", 999999.9, "BILLIONS", 999.9)

my_account.deposit(999.9)
print("Current balance:", my_account.check_balance())

my_account.withdraw(200.0)
print("Current balance:", my_account.check_balance())
```
